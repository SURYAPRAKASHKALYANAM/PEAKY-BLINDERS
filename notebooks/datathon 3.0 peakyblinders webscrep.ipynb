# extract prefix to webscrap(domain)

```python
df["Gist Link"]=df["Gist Link"].apply(lambda x:str(x))
```

```python
# Function to extract the desired format from a URL
def extract_website(url):
    try:
        result = re.match(r'(https?://[^/]+/)', url)
        if result:
            return result.group(1)
        else:
            return ''
    except:
        pass
df['extracted_website'] = df["Gist Link"].apply(lambda x:extract_website(x))
```

# web scraping

```python
import requests
from bs4 import BeautifulSoup
def whitehouse(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    sections = soup.find_all('section', class_='body-content')
    res = ""
    for section in sections:
        p_elements = section.find_all('p')
        for p_element in p_elements:
            res += p_element.get_text() + "\n"  
    return res
```

```python
def obamawhitehouse(url_given):
    url_given = requests.get(url_given)
    soup = BeautifulSoup(url_given.content, 'html.parser')
    cards = soup.find_all('div', class_='field-item even')
    text=""
    for card in cards:
        p_elements = card.find_all('p')
        for p_elem in p_elements:
            text += p_elem.get_text() + "\n"  
    return text
```

```python
def govuk(url_given):
    url_given = requests.get(url_given)
    soup = BeautifulSoup(url_given.content, 'html.parser')
    cards = soup.find_all('div', class_='responsive-bottom-margin')
    text=""
    for card in cards:
        p_elements = card.find_all('p')
        for p_elem in p_elements:
            text += p_elem.get_text() + "\n"  
    return text
```

```python
def trumpwhitehouse(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    p_tags = soup.find_all('p')
    paragraphs = []
    
    for p in p_tags:
        paragraphs.append(p.get_text())
    
    formatted_text = ''.join(paragraphs)
    
    return paragraphs
```
